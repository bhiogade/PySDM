```python
from PySDM_examples.Jensen_and_Nugent_2016 import Settings, Simulation
from open_atmos_jupyter_utils import show_plot
from PySDM_examples.Jensen_and_Nugent_2016.plotting import figure
from PySDM.physics import si

settings = Settings(aerosol="modified polluted", cloud_type="Cu")
simulation = Simulation(settings, gccn=True)
```

```python
output = simulation.run(n_steps=int(1800 * si.m / settings.vertical_velocity / settings.dt))
```

```python
figure(
    output=output,
    settings=settings,
    simulation=simulation,
    drop_ids=(-80, -37, -32, -27, -22, -17, -12, -7, -3, -1),
    xlim_r_um=(0, 60),
    xlim_S_percent=(-1.5, 1.5)
)
show_plot("Fig_6.pdf")
```
